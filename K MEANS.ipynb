```python
# K MEANS
from sklearn.datasets import load_iris # Iris data set
iris=load_iris()
iris
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as  plt

X,y=iris['data'],iris['target'] 

cluster_li,distance_li=[],[]
```

```python
for n_clusters in range(1,10):

    print('test n_cluster=%d ...'%n_clusters)
    kmean=KMeans(n_clusters=n_clusters) 
    kmean.fit(X) 

    distance_li.append(kmean.inertia_)
    cluster_li.append(n_clusters)
```

```python
plt.figure(figsize=(20,8))
plt.title('Elbow Heuristic Figure')
plt.plot(cluster_li,distance_li)
plt.show()     
```

```python
# As can see, when the critical point 3 is reached, the degree of 
# distortion has been greatly improved, and then slowly decreases.
# So, the assumed number of clusters on the data is 3.
```
